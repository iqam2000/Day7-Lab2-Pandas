```python
#Q13: Rename all dataset's columns
insta = insta.rename(columns={"pos":"Num_posts",
                      "flg":"Num_following",
                      "bl":"Biography_length",
                      "pic":"Picture_availability",
                      "lin":"Link_availability",
                      "cl":"Average_caption_length",
                      "cz":"Caption_zero",
                      "ni":"Non_image_percentage",
                      "erl":"Engagement_rate_like",
                      "erc":"Engagement_rate_comment",
                      "lt":"Location_tag_percentage",
                      "hc":"Average hashtag count",
                      "pr":"Promotional keywords",
                      "fo":"Followers keywords",
                      "cs":"Cosine similarity",
                      "pi":"Post interval",
                      "class":"real_fake"})

insta
```

```python
#Q14: Change the class's values to real and fake
insta["real_fake"] = insta["real_fake"].str.replace("r", "real")
insta["real_fake"] = insta["real_fake"].str.replace("f", "fake")
insta 
```

```python
#Q15: Print the total number of each fake accounts and real accounts
print(len(insta[insta["real_fake"]=="real"]))
print(len(insta[insta["real_fake"]=="fake"]))
```
